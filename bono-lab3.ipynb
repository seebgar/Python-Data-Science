```python

## graficar la velocidad y aceleracion
## T = 3
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc 

A = 5                               # Amplitud máxima del movimiento
T = 8                               # Periodo del movimiento (El tiempo que tarda en dar una oscilación completa)
w = 2*np.pi / T                     # Frecuencia Angular (Número de periodos comprendidos en 2·π segundos)
t = np.linspace(0,20,1000)          # Vector de tiempos

x = A * np.sin(w*t + np.pi/2)                 # Ecuación de posición del m.a.s

plt.style.use('seaborn-white')
plt.title('Gráfica de posición (m.a.s)')
plt.xlabel('t(s)')
plt.ylabel('Amplitud (m)')

plt.plot(t, x, lw=4, color='c')

def func(t):
    return A * np.sin(w*t + np.pi/2) 

d = misc.derivative(func, t) 
plt.plot(t, d, lw=4, color='r')

ds = misc.derivative(func, d) 
plt.plot(t, ds, lw=4, color='g')
```
