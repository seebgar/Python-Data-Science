# Numerical Differentiation

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

'''
3 Main Formulas
* forward => [f(a+h) - f(a)] / h
* backward => [f(a) - f(a-h)] / h
* centrak => average of above => [f(a+h) - f(a-h)] / 2h
'''

def derivative(f, a, h=0.01):
    '''
    Compute the difference formula for f'(a) with step size h.

    Parameters
    ----------
    f : function
        Vectorized function of one variable
    a : number
        Compute derivative at x = a
    h : number
        Step size in difference formula

    Returns
    -------
    float
        Difference formula:
            central: f(a+h) - f(a-h))/2h
    '''
    return (f(a + h) - f(a - h))/(2*h)
```

```python
# derivative of sin(x)

x = np.linspace(0,5*np.pi,100)
dydx = derivative(np.sin,x)

dYdx = np.cos(x)

plt.figure(figsize=(12,5))
plt.plot(x,dydx,'r.',label='Central Difference')
plt.plot(x,dYdx,'b',label='True Value')

plt.title('Central Difference Derivative of y = cos(x)')
plt.legend(loc='best')
plt.show()
```

```python
# derivative of custom fuction

x = np.linspace(0,6,100)
f = lambda x: ((4*x**2 + 2*x + 1)/(x + 2*np.exp(x)))**x
dydx = derivative(f,x)
y = f(x)

plt.figure(figsize=(12,5))
plt.plot(x,y,label='y=f(x)')
plt.plot(x,dydx,label="Central Difference y=f'(x)")

plt.legend()
plt.grid(True)
plt.show()
```

# Bisection Method

The slgorithm applies to any **continuos** function *f(x)* on an interval *[a, b]* where the value of the function *f(x)* **changes sign** from *a* to *b*.
<br/>
The idea is simple: divide the interval in two, a solution must exist within one subinterval, select the subinterval where the sign of *f(x)* changes and repeat.


```python

def bisection(f, a, b, N=1000, tlr_x=1e-05, tlr_y=1e-05 ):
    '''
    Approximate solution of f(x)=0 on interval [a,b] by bisection method.

    Parameters
    ----------
    f : function
        The function for which we are trying to approximate a solution f(x)=0.
    a,b : numbers
        The interval in which to search for a solution. The function returns
        None if f(a)*f(b) >= 0 since a solution is not guaranteed.
    N : (positive) integer
        The number of iterations to implement.

    Returns
    -------
    x_N : number
        The midpoint of the Nth interval computed by the bisection method. The
        initial interval [a_0,b_0] is given by [a,b]. If f(m_n) == 0 for some
        midpoint m_n = (a_n + b_n)/2, then the function returns this solution.
        If all signs of values f(a_n), f(b_n) and f(m_n) are the same at any
        iteration, the bisection method fails and return None.

    Examples
    --------
    >>> f = lambda x: x**2 - x - 1
    >>> bisection(f,1,2,25)
    1.618033990263939
    >>> f = lambda x: (2*x - 1)*(x - 3)
    >>> bisection(f,0,1,10)
    0.5
    '''
    if f(a) * f(b) >= 0:
        print("Bisection method fails (The sign of the function must change).")
        return None
    
    roots = []
    a_n = a
    b_n = b
    prev_root = b_n
    
    for n in range(1, N+1):
        m_n = (a_n + b_n)/2
        f_m_n = f(m_n)
        roots.append(m_n)
        
        # Tolerancias 10**-5
        if np.abs(m_n - prev_root) <= tlr_x:
            break
        if np.abs(f_m_n) <= tlr_y:
            break
        
        if f(a_n) * f_m_n < 0:
            a_n = a_n
            b_n = m_n
        elif f(b_n) * f_m_n < 0:
            a_n = m_n
            b_n = b_n
        elif f_m_n == 0:
            return m_n, roots
        else:
            print("Bisection method fails.")
            return None
        
        prev_root = m_n
       
    m_n = (a_n + b_n)/2
    return  m_n, roots
```

```python
# golden ratio phi

f = lambda x: x**2 - x - 1
phi = bisection(f,1,2,25)
print(phi[0])

# ejemplo mrk

g = lambda x: -(np.sqrt(3.0*x)**(2.0/5.0)) + (x**3.0) * np.cos(3.0*x) + (4.0*(x**2.0)) - 7
raiz = bisection(g, 1, 2)
print(raiz)
```

```python
1.618033990263939
```

# Newton's Method

Newton's method is a root finding method that uses linear approximation. In particular, we guess a solution **xo** of the equation **f(x)=0**, compute the linear approximation of **f(x)** at **x0** and then find the **x**-intercept of the linear approximation.
<br/>
Let **f(x)** be a differentiable function. If **x0** is near a solution of **f(x)=0** then we can approximate **f(x)** by the tangent line at **x0** and compute the **x**-intercept of the tangent line.

```python

def newton(f, Df, x, N=1000, tlr_x=(10**-10), tlr_y=(10**-10)):
    '''
    Approximate solution of f(x)=0 by Newton's method.

    Parameters
    ----------
    f : function
        Function for which we are searching for a solution f(x)=0.
    Df : function
        Derivative of f(x).
    x0 : number
        Initial guess for a solution f(x)=0.
    epsilon : number
        Stopping criteria is abs(f(x)) < epsilon.
    max_iter : integer
        Maximum number of iterations of Newton's method.

    Returns
    -------
    xn : number
        Implement Newton's method: compute the linear approximation
        of f(x) at xn and find x intercept by the formula
            x = xn - f(xn)/Df(xn)
        Continue until abs(f(xn)) < epsilon and return xn.
        If Df(xn) == 0, return None. If the number of iterations
        exceeds max_iter, then return None.

    Examples
    --------
    >>> f = lambda x: x**2 - x - 1
    >>> Df = lambda x: 2*x - 1
    >>> newton(f,Df,1,1e-8,10)
    Found solution after 5 iterations.
    1.618033988749989
    '''
    roots = []
    while True:
        x1 = x - f(x) / Df(x)
        roots.append(x1)
        if np.abs(x1 - x) < tlr_x:
            break
        if np.abs(f(x1)) < tlr_y:
            break
        x = x1
    return x, roots
```

```python
# supergolden ratio
'''
p = lambda x: x**3 - x**2 - 1
Dp = lambda x: 3*x**2 - 2*x
approx = newton(p, Dp, 1)
print(approx)
'''

# ejemplo mrk

f = lambda x: -(np.sqrt(3.0*x)**(2.0/5.0)) + (x**3.0) * np.cos(3.0*x) + (4.0*(x**2.0)) - 7
df = lambda x: (((-3.0**(1.0/5.0)) / 5) * (x**(-4.0/5.0))) - (3.0* (x**3) * np.sin(3.0*x)) + (3.0* (x**2) * np.cos(3.0*x)) + (8.0*x)
raiz = newton(f, df, 2.0)
print(raiz)
```

```python
1.5030624561642578
```

# Secant Method

The secant method is very similar to the bisection method except instead of dividing each interval by choosing the midpoint **the secant method divides each interval by the secant line connecting the endpoints**. The secant method always converges to a root of **f(x)=0** provided that **f(x)** is continuous on **[a,b]** and **f(a)f(b)<0** .

```python

def secant(f, a, b, N=1000, tlr_x=1e-05, tlr_y=1e-05):
    '''
    Approximate solution of f(x)=0 on interval [a,b] by the secant method.

    Parameters
    ----------
    f : function
        The function for which we are trying to approximate a solution f(x)=0.
    a,b : numbers
        The interval in which to search for a solution. The function returns
        None if f(a)*f(b) >= 0 since a solution is not guaranteed.
    N : (positive) integer
        The number of iterations to implement.

    Returns
    -------
    m_N : number
        The x intercept of the secant line on the the Nth interval
            m_n = a_n - f(a_n)*(b_n - a_n)/(f(b_n) - f(a_n))
        The initial interval [a_0,b_0] is given by [a,b]. If f(m_n) == 0
        for some intercept m_n then the function returns this solution.
        If all signs of values f(a_n), f(b_n) and f(m_n) are the same at any
        iterations, the secant method fails and return None.

    Examples
    --------
    >>> f = lambda x: x**2 - x - 1
    >>> secant(f,1,2,5)
    1.6180257510729614
    '''
    roots = []
    a_n = a
    b_n = b
    prev_root = b_n
    
    for n in range(1, N+1):
        m_n = b_n - ((f(b_n) * (b_n - a_n)) / (f(b_n) - f((a_n))) )
        f_m_n = f(m_n)
        roots.append(m_n)
        
        # Tolerancias 10**-5
        if np.abs(m_n - prev_root) <= tlr_x:
            break
        if np.abs(f_m_n) <= tlr_y:
            break
        
        a_n = b_n
        b_n = m_n
        prev_root = m_n
        
    m_n = b_n - ((f(b_n) * (b_n - a_n)) / (f(b_n) - f((a_n))) )
    return m_n, roots
```

```python
# supergolden ratio

f = lambda x: x**3 - x**2 - 1
r = secant(f, 1, 2, 20)
print(r[0])

# ejemplo mrk

g = lambda x: -(np.sqrt(3.0*x)**(2.0/5.0)) + (x**3.0) * np.cos(3.0*x) + (4.0*(x**2.0)) - 7
raiz = secant(g, 2, 1.9)
print(raiz)
```

# False position

Lo mismo que Bisección pero cambia el criterio para determinar m_n. M_n ahora se tomará como el punto que corta el eje x que se encuentra sobre la recta que une los puntos (a, f(a)) y (b, f(b)).

```python

def regula_falsi(f, a, b, N=1000, tlr_x=1e-05, tlr_y=1e-05):
    '''
    Approximate solution of f(x)=0 on interval [a,b] by the false position method.
    
    Parameters
    ----------
    f : function
        The function for which we are trying to approximate a solution f(x)=0.
    a,b : numbers
        The interval in which to search for a solution. The function returns
        None if f(a)*f(b) >= 0 since a solution is not guaranteed.
    N : (positive) integer
        The number of iterations to implement.

    Returns
    -------
    m_N : number
        The x intercept of the secant line on the the Nth interval
            m_n = (a_n * f(b_n) - b_n * f(a_n)) / (f(b_n) - f(a_n))
        The initial interval [a_0,b_0] is given by [a,b]. If f(m_n) == 0
        for some intercept m_n then the function returns this solution.
        If all signs of values f(a_n), f(b_n) and f(m_n) are the same at any
        iterations, the false position method fails and return None.

    '''
    
    if f(a)*f(b) >= 0:
        print("False Position method fails (The sign of the function must change).")
        return None
    
    roots = []
    a_n = a
    b_n = b
    prev_root = b_n
    
    for n in range(1, N+1):
        # m_n = (a_n * f(b_n) - b_n * f(a_n)) / (f(b_n) - f(a_n))
        m_n = b_n - ((f(b_n) * (b_n - a_n)) / (f(b_n) - f((a_n))) )
        f_m_n = f(m_n)
        roots.append(m_n)
        
        # Tolerancias 10**-5
        if np.abs(m_n - prev_root) <= tlr_x:
            break
        if np.abs(f_m_n) <= tlr_y:
            break
        
        if f(a_n) * f_m_n < 0:
            a_n = a_n
            b_n = m_n
        elif f(b_n) * f_m_n < 0:
            a_n = m_n
            b_n = b_n
        elif f_m_n == 0:
            return m_n, roots
        else:
            print("False Position method fails.")
            return None
        

        
        prev_root = m_n
       
    m_n = b_n - ((f(b_n) * (b_n - a_n)) / (f(b_n) - f((a_n))) )
    return m_n, roots
```

```python
# supergolden ratio

f = lambda x: x**3 - x**2 + 2
r = regula_falsi(f, -200, 300)
print(r[0])

# example

f2 = lambda x: x**3 - 5*x - 9
r2 = regula_falsi(f2, 2, 3)
print(r2[0])

# ejemplo mrk

g = lambda x: -(np.sqrt(3.0*x)**(2.0/5.0)) + (x**3.0) * np.cos(3.0*x) + (4.0*(x**2.0)) - 7
raiz = regula_falsi(g, 1, 2)
print(raiz)
```

# Punto Fijo
Se busca una funcón cualquiera g(x) tal que al evaluar **g** en **x** se obtiene el valor **x** => **g(x) = x**.
<br/>
De esta forma se halla la solución a **f(x) = g(x) - x = 0**
<br/>
m_n = g(b_n)

```python

def fixed_point(f, b, g, N=1000, tlr_x=1e-05, tlr_y=1e-05):
    '''
    Approximate solution of f(x)=0 on interval [a,b] by the fixed point method.
    
    Parameters
    ----------
    f : function
        The function for which we are trying to approximate a solution f(x)=0.
    b : number
        fixed point near the root
    N : (positive) integer
        The number of iterations to implement.

    Returns
    -------
    m_N : number
    '''
    
    roots = []
    b_n = b
    prev_root = b_n
    
    for n in range(1, N+1):
        m_n = g(b_n)
        f_m_n = f(m_n)
        roots.append(m_n)
        
        # Tolerancias 10**-5
        if np.abs(m_n - prev_root) <= tlr_x:
            break
        if np.abs(f_m_n) <= tlr_y:
            break
        
        b_n = m_n
        prev_root = m_n
        
    return g(b_n), roots
```

```python

# ejemplo mrk

g = lambda x: ((1.0/4.0) * (np.sqrt(3.0 * x) ** (2.0/5.0) - (x**3.0) * np.cos(3.0*x) + 7)) ** (0.5)
f = lambda x: -(np.sqrt(3.0*x)**(2.0/5.0)) + (x**3.0) * np.cos(3.0*x) + (4.0*(x**2.0)) - 7
raiz = fixed_point(f, 2.0, g)
print(raiz)
```

# Convergence Rate

Una tasa de convergencia es alta si se necesitan menos iteraciones para converger - encontrar la raíz.

```python
import scipy.optimize as opt

def rate(x_vals, f, x):
    '''
    Convergence rate for a given x_values roots and the exact root.
    '''
    x_exact = opt.fsolve(f, x)
    e = [ np.abs(_x - x_exact) or 1 for _x in x_vals ]
    q = [ np.log10(e[n+1]/e[n]) / np.log10(e[n]/e[n-1]) for n in range(1, len(e) - 1, 1)]
    return q
```

```python
f = lambda x: -(np.sqrt(3.0*x)**(2.0/5.0)) + (x**3.0) * np.cos(3.0*x) + (4.0*(x**2.0)) - 7
df = lambda x: (((-3.0**(1.0/5.0)) / 5) * (x**(-4.0/5.0))) - (3.0* (x**3) * np.sin(3.0*x)) + (3.0* (x**2) * np.cos(3.0*x)) + (8.0*x)

## biseccion
print('''
    Bisection
''')
raiz, roots = bisection(f, 1, 2)
tasa = rate(roots, f, 2)
print(tasa[-2:-1])

## newton
print('''
    Newton
''')
raiz, roots = newton(f, df, 2.0)
tasa = rate(roots, f, 2)
print(tasa[-2:-1])
```
