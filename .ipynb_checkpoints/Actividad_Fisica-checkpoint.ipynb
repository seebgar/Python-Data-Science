```python
##

"""
Parte gráfica (1)
"""

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

U0 = 1.60 * (10**(-21))
R0 = 3.82 * (10**(-10))
    
def u(r):
    return U0 * (((R0/r) ** 12) - (2 * ((R0/r) ** 6))) 

x = np.arange((R0/2), (R0*2), (R0/1000) )
y = u(x)

plt.figure(figsize=(12,5))
plt.plot(x,y,label='y=u(r)')
plt.plot(x,[0 for i in y],'r--') # EJE X

plt.legend()
plt.grid(True)
plt.show()
```

```python
##

"""
Parte gráfica (2)
"""

dru = lambda r: (-2 * (r**6) + (R0**6))/(12 * (r**5) * U0)

x = np.arange((R0/2), (R0*2), (R0/1000) )
y = dru( x)

plt.figure(figsize=(12,5))
plt.plot(x,y,label='y=u(r)')
plt.plot(x,[0 for i in y],'r--') # EJE X

plt.legend()
plt.grid(True)
plt.show()
```

```python
##

"""
Parte computacional (1)
"""

def bisection(f, a, b, N=1000, tlr_x=1e-05, tlr_y=1e-05 ):
    if f(a) * f(b) >= 0:
        print("El signo de la función debe cambiar dentro del intervalo [a,b]")
        return None
    
    roots = []
    a_n = a
    b_n = b
    prev_root = b_n
    
    for n in range(1, N+1):
        m_n = (a_n + b_n)/2
        f_m_n = f(m_n)
        roots.append(m_n)
        
        # Tolerancias 10**-5
        if np.abs(m_n - prev_root) <= tlr_x:
            break
        if np.abs(f_m_n) <= tlr_y:
            break
        
        if f(a_n) * f_m_n < 0:
            a_n = a_n
            b_n = m_n
        elif f(b_n) * f_m_n < 0:
            a_n = m_n
            b_n = b_n
        elif f_m_n == 0:
            return m_n, roots
        else:
            print("Sin solución.")
            return None, roots
        
        prev_root = m_n
       
    m_n = (a_n + b_n)/2
    return  m_n, roots

# biseccion
raiz, roots = bisection(u, R0/2, R0*1.5, tlr_x=10**(-30), tlr_y=10**(-30))
ur = u(raiz)

print(f"""  
    Bisección
    No. iteraciones:\t {len(roots)}
    Raíz xr:\t\t {raiz}
    u(xr):\t\t {ur}
""")
```

```python
##

from tabulate import tabulate

"""
Parte computacional (2)
"""

def fixed_point(f, g, b, N=1000, tlr_x=1e-05, tlr_y=1e-05):
    roots = []
    b_n = b
    prev_root = b_n
    
    for n in range(1, N+1):
        m_n = g(b_n)
        f_m_n = f(m_n)
        roots.append(m_n)
        
        # Tolerancias 10**-5
        if np.abs(m_n - prev_root) <= tlr_x:
            break
        if np.abs(f_m_n) <= tlr_y:
            break
        
        b_n = m_n
        prev_root = m_n
        
    return g(b_n), roots

g = lambda r: (( (r**6) * (R0**12) * U0 ) / ( 2.0 * (R0**6) * U0 )) ** (1/12)
p = R0 * 1.5

raiz_i, roots_i = fixed_point(u, g, p, tlr_x=10**(-30), tlr_y=10**(-30))

tb = []
for i in range(len(roots_i)):
    tb.append([i, roots_i[i], g(roots_i[i]), roots_i[i]])

tb_i = tabulate(tb,
    headers=['Iteración', 'nr', 'g(nr)', '|nr_i - nr_i-1|']
)

print(f"""
Punto fijo

Iteraciones: {len(roots_i)}
Raís mr: {raiz_i}
u(mr): {u(raiz_i)}

{tb_i}
""")
```

```python
##

"""
Bono
"""

def derivative(f, a, h=0.01):
    return (f(a + h) - f(a - h))/(2*h)

def newton(f, x, tlr_x=1e-05, tlr_y=1e-05):
    roots = []
    while True:
        Dfx = derivative(f, x)
        if Dfx == 0:
            print("Derivada igual a 0.")
            return None, roots
        
        x1 = x - (f(x) / Dfx)
        roots.append(x1)
        
        # Tolerancias 10**-5
        if np.abs(x1 - x) < tlr_x:
            break
        if np.abs(f(x1)) < tlr_y:
            break
            
        x = x1
        
    return x, roots

# newton
raiz_i_n, roots_i_n = newton(u, R0*0.9 )

tb = []
for i in range(len(roots_i_n)):
    tb.append([i, roots_i_n[i], u(roots_i_n[i]), roots_i_n[i]])

tb_i = tabulate(tb,
    headers=['Iteración', 'nr', 'u(nr)', '|nr_i - nr_i-1|']
)

print(f"""
Newton

Iteraciones: {len(roots_i_n)}
Raís mr: {raiz_i_n}
u(mr): {u(raiz_i_n)}

{tb_i}
""")
```
