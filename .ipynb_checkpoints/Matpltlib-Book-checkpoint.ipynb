```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
### inline => static image
### noebook => interactive 

# plt.style.use('classic')

fig = plt.figure()
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')

# plt.show() # should be used only once per script at the end
```

```python
# Object oriented interface
# grid of plots
# ax will be an array of two Axes objects

fig, ax = plt.subplots(2)

# call plot() on the appropiate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

# line plot

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-notebook') # seaborn-muted | dark_background

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)

# colors = rgbcmyk | html | hex
# linestyle = solid - | dashed -- | dashdot -. | dotted :

ax.plot(x, np.sin(x), color='#34a8eb', linestyle='--', label='sin(x)') # blue
ax.plot(x, np.cos(x), color="#bd34eb", linestyle='-', label='cos(x)') # purple

# axis limits
# plt.xlim(-1, 11)
# plt.ylim(-1.5, 1.5)

plt.axis([-1, 11, -1.5, 1.5])
plt.axis('equal')

# labels

plt.title('Sine & Cosine')
plt.xlabel("x")
plt.ylabel('f(x)')
plt.legend() # ax.legend()

## 
#ax.set(
#    xlim=(0,10),
#    ylim=(-2,2),
#    xlabel="x",
#    ylabel="f(x)",
#    title="Plot"
#)
# ax.legend()
```

# scatter plot

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-notebook') # seaborn-muted | dark_background

x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.scatter(x, y, marker='o')
```

```python
##

rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
```

# error bar

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 30)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(30)

plt.errorbar(x, y, yerr=dy, fmt='o', color='black', ecolor='lightgray', elinewidth=3, capsize=0)
```
