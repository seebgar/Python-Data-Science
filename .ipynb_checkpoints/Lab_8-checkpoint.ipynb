### Ejercicio 1
Realice un programa en Python que, a partir de un conjunto de puntos observados (x, y), implemente la interpolación polinomial usando el método matricial (matriz de Vandermonde). Adicionalmente, el programa debe estimar, para una serie de valores xnew, los correspondientes valores interpolados ynew.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def get_coeficientes(x_i, y_i):
    """
    Caclculo de coeficientes a apartir de un conjunto
    de valores de entrada x_i, y_i
    """
    N = np.size(x_i)
    A = np.zeros([N, N]);
    b = np.zeros([N, 1])
    
    for i in range(N):
        for j in range(N):
            A[i, j] = x_i[i] ** (N - j - 1)
        b[i] = y_i[i]
    
    c_i = np.linalg.solve(A, b)
    return c_i

def get_Yk(c_i, x_k):
    """
    Estimación de valores Yk a partir de un conjunto de 
    valores deseados Xk y los coeficientes de los puntos 
    X_i iniciales conocidos.
    """
    N = np.size(c_i)
    Nxk = np.size(x_k)
    Yk = np.zeros(Nxk)
    
    for i in range(Nxk):
        for j in range(N):
            Yk[i] += (c_i[j] * x_k[i] ** (N - j - 1))
    
    return Yk

def get_Interpolacion_Polinomial(x_i, y_i, x_k, xlim_max=80, wg=True):
    """
    Interpolación polinomial a partir de un conjunto de valores
    conocidos x_i, y_i; y un conjunto de valores a conocer 
    sus valores x_k
    Se retornan las estimaciones y_k y se grafica el resultado.
    """
    c_i = get_coeficientes(x_i, y_i)
    y_k = get_Yk(c_i, x_k)
    
    if wg:
        # grafica
        x_p = np.arange(x_i[0], x_i[-1], 0.01)
        y_p = get_Yk(c_i, x_p)
        plt.figure()
        plt.figure(figsize=(12,5))
        plt.plot(x_p, y_p, 'b') # polinomio obtenido
        plt.plot(x_i, y_i, 'sr') # puntos iniciales
        plt.plot(x_i, y_i, '--', color='0.5') # recta entre puntos iniciales
        plt.plot(x_k, y_k, 'sg') # puntos estimados
        plt.xlim([0.0, xlim_max])
        dis = (np.max(y_p) - np.min(y_p)) * 0.1
        plt.ylim([np.min(y_p) - dis, np.max(y_p) + dis])
        plt.xlabel("X")
        plt.ylabel("Y")
        plt.grid(1)
    
    return y_k

# defincion de puntos x,y conocidos
Xi = np.array([13, 16, 17, 23, 28, 33])
Yi = np.array([30, 92, 21, 159, 94, 236])
# valores x a encontrar solucion f(x)
Xk = np.array([15, 19, 21, 25, 27, 32])

Yk = get_Interpolacion_Polinomial(Xi, Yi, Xk, xlim_max=35)

print("Xk\t{:^35s}\t".format("Yk"))
for i in range(np.size(Yk)):
    print("{:d}\t{: 25.10f}\t".format(Xk[i], Yk[i]))
    
```

### Ejercicio 2
Realice un programa en Python que, a partir de un conjunto de puntos observados (x, y), implemente la interpolación polinomial usando el método de Lagrange. El programa debe estimar, para una serie de valores xnew, los correspondientes valores interpolados ynew.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def get_Yk_Lagrange(x_i, y_i, x_k):
    Nxk = np.size(x_k)
    Nyi = np.size(y_i)
    y_k = np.zeros(Nxk)
    
    for k in range(Nxk):
        for i in range(Nyi):
            m_aux = 1
            for j in range(Nyi):
                if i == j:
                    continue
                m_aux *= ((x_k[k] - x_i[j]) / (x_i[i] - x_i[j]))
            y_k[k] += y_i[i] * m_aux
    
    return y_k

def get_Interpolacion_Lagrange(x_i, y_i, x_k, xlim_max=80, wg=True):
    y_k = get_Yk_Lagrange(x_i, y_i, x_k)
    c_i = get_coeficientes(x_i, y_i)
    
    if wg:
        # grafica
        x_p = np.arange(x_i[0], x_i[-1], 0.01)
        y_p = get_Yk(c_i, x_p)
        plt.figure()
        plt.figure(figsize=(12,5))
        plt.plot(x_p, y_p, 'b') # polinomio obtenido
        plt.plot(x_i, y_i, 'sr') # puntos iniciales
        plt.plot(x_i, y_i, '--', color='0.5') # recta entre puntos iniciales
        plt.plot(x_k, y_k, 'sg') # puntos estimados
        plt.xlim([0.0, xlim_max])
        dis = (np.max(y_p) - np.min(y_p)) * 0.1
        plt.ylim([np.min(y_p) - dis, np.max(y_p) + dis])
        plt.xlabel("X")
        plt.ylabel("Y")
        plt.grid(1)

    return y_k

# defincion de puntos x,y conocidos
Xi = np.array([13, 16, 17, 23, 28, 33])
Yi = np.array([30, 92, 21, 159, 94, 236])
# valores x a encontrar solucion f(x)
Xk = np.array([15, 19, 21, 25, 27, 32])

Yk = get_Interpolacion_Lagrange(Xi, Yi, Xk, xlim_max=35)

print("Xk\t{:^35s}\t".format("Yk"))
for i in range(np.size(Xk)):
    print("{:d}\t{: 25.10f}\t".format(Xk[i], Yk[i] ))
    
```

### Ejercicio 3
En los archivos x_obs.bin y y_obs.bin se encuentran una serie de puntos observados (x, y) almacenados en formato binario de tipo double. A partir de los procedimientos implentados en los puntos 1 y 2, complete la siguiente tabla con los valores de ynew correspondientes a cada uno de los valores presentados en la columna xnew. Complete adicionalmente la última columna con los correspondientes valores ynew estimados por la función nativa de interpolación polinomial de numpy.

```python
import numpy as np
import matplotlib.pyplot as plt
import struct
%matplotlib inline
from tabulate import tabulate

x_file_name = "x_obs.bin"
y_file_name = "y_obs.bin"

Xi = np.array([])
Yi = np.array([])
Xk = np.array(
    [78.12, 0.98, 67.59, 8.69, 55.69, 48.12, 13.24, 97.56, 25.69, 1.26]
, dtype='float')

with open(x_file_name, 'rb') as f:
    x = f.read()
    Xi = np.array(struct.unpack("d"*int(len(x)/8), x)) 
    
with open(y_file_name, 'rb') as f:
    y = f.read()
    Yi = np.array(struct.unpack("d"*int(len(y)/8), y)) 
    

# interpolacion polinomial
Yk_ip = get_Interpolacion_Polinomial(Xi, Yi, Xk, wg=False)

# interpolacion lagrange
Yk_lg = get_Interpolacion_Lagrange(Xi, Yi, Xk, wg=False)

# numpy
CiPoly = np.polyfit(Xi, Yi, 1)
YkPoly = np.polyval(CiPoly, Xk)

# return
tb_itr = []
for i in range(np.size(Xk)):
    row = [Xk[i], Yk_ip[i], Yk_lg[i], YkPoly[i] ]
    tb_itr.append(row)

tb = tabulate( tb_itr,
    headers=['Xk', 'Yk Polinomial', 'Yk Lagrange', 'Yk Numpy']
)

print(f"""
{tb}
""")
```

### Ejercicio 5
A parir del mismo conjunto de datos observados presentados en los archivos del punto 3, utilice ahora el método de interpolación por splines cúbicos de la librería scipy para completar la tabla presentada a continuación,

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as interpol
import struct
%matplotlib inline
from tabulate import tabulate

x_file_name = "x_obs.bin"
y_file_name = "y_obs.bin"

Xi = []
Yi = []
Xk = np.array(
    [78.12, 0.98, 67.59, 8.69, 55.69, 48.12, 13.24, 97.56, 25.69, 1.26]
)

with open(x_file_name, 'rb') as f:
    x = f.read()
    Xi = np.array(struct.unpack("d"*int(len(x)/8), x)) 
    
with open(y_file_name, 'rb') as f:
    y = f.read()
    Yi = np.array(struct.unpack("d"*int(len(y)/8), y)) 

# Splines Natural :: segunda derivada a cero en los puntos extremos
f1Nat = interpol.CubicSpline(Xi, Yi, bc_type="natural")
# Splines Not - a - knot :: igualdad de terceras drivadas en el segundo y penultimo punto. (se necesitan al menos 4 puntos)
f1Nak = interpol.CubicSpline(Xi, Yi, bc_type="not-a-knot")
# Splines Clamped, con valor k=0 :: primera derivada en los puntos extremos igual a un valor definido k
f1Clam = interpol.CubicSpline(Xi, Yi, bc_type="clamped")

"""
Gráfica de polinomios resultantes
"""

# valores estimados de Y usando cada tipo de spline
YkNat = f1Nat(Xk)
YkNak = f1Nak(Xk)
YkClam = f1Clam(Xk)

# rango
Xp = np.arange(Xi[0], Xi[-1] - 0.001, 0.001)

plt.figure()
plt.figure(figsize=(14,7))
plt.plot(Xi, Yi, 'sr')
plt.plot(Xp, f1Nat(Xp), '--b')
plt.plot(Xp, f1Nak(Xp), '--m')
plt.plot(Xp, f1Clam(Xp), '--g')
plt.xlim([0, 100])
plt.ylim([-150, 300])
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["Conocidos", "Natural", "Not-a-Knot", "Clamped"])
plt.grid(1)



# imprimir los valores y comparar
print("Xk\t{:^15s}\t".format("Yk Natural"))
for i in range(np.size(Xk)):
    print("{:0.2f}\t{: 15.10f}\t".format(Xk[i], YkNat[i] ))
    
```
