```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 2, 6, 14, 30, 43, 75]

# axis labels
plt.ylabel('Speed', fontsize=12)
plt.xlabel('Time', fontsize=12)

# title
plt.title('Speed v Time')

# display
plt.plot(x, y, 'bo-')
plt.show()
```

```python
# setup data
riding = ((17, 18, 21, 22, 19, 21, 25, 22, 25, 24),
          (3, 6, 3.5, 4, 5, 6.3, 4.5, 5, 4.5, 4))
swimming = ((17, 18, 20, 19, 22, 21, 23, 19, 21, 24),
            (8, 9, 7, 10, 7.5, 9, 8, 7, 8.5, 9))
sailing = ((31, 28, 29, 36, 27, 32, 34, 35, 33, 39),
           (4, 6.3, 6, 3, 5, 7.5, 2, 5, 7, 4))

# plot data
plt.scatter(x=riding[0], y=riding[1], c='red', marker='o', label='riding')
plt.scatter(x=swimming[0], y=swimming[1],
            c='green', marker='^', label='swimming')
plt.scatter(x=sailing[0], y=sailing[1],
            c='blue', marker='*', label='sailing')

# config
plt.xlabel("Age")
plt.ylabel("Hours")
plt.title("Activities Scatter Graph")
plt.legend()

# display
plt.show()
```

```python
ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```python
from scipy.integrate import odeint
from numpy import cos, sin


n = np.array([0,1,2,3,4,5])
xx = np.linspace(-0.75, 1., 100)

fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

```
