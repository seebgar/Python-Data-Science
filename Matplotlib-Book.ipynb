```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
### inline => static image
### noebook => interactive 

# plt.style.use('classic')

fig = plt.figure()
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')

# plt.show() # should be used only once per script at the end
```

# line plot

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-notebook') # seaborn-muted | dark_background

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)

# colors = rgbcmyk | html | hex
# linestyle = solid - | dashed -- | dashdot -. | dotted :

ax.plot(x, np.sin(x), color='#34a8eb', linestyle='--', label='sin(x)') # blue
ax.plot(x, np.cos(x), color="#bd34eb", linestyle='-', label='cos(x)') # purple

# axis limits
# plt.xlim(-1, 11)
# plt.ylim(-1.5, 1.5)

plt.axis([-1, 11, -1.5, 1.5])
plt.axis('equal')

# labels

plt.title('Sine & Cosine')
plt.xlabel("x")
plt.ylabel('f(x)')
plt.legend() # ax.legend()

## 
#ax.set(
#    xlim=(0,10),
#    ylim=(-2,2),
#    xlabel="x",
#    ylabel="f(x)",
#    title="Plot"
#)
# ax.legend()
```

# error bar

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 30)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(30)

plt.errorbar(x, y, yerr=dy, fmt='o', color='black', ecolor='lightgray', elinewidth=3, capsize=0)
```

# histograms

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-notebook') # seaborn-muted | dark_background

data = np.random.randn(1000)
particiones = 10
kwargs = dict(bins=particiones, alpha=0.9, histtype='stepfilled', color='#34a8eb', edgecolor='none')

plt.hist(data, **kwargs)
plt.show()

cantidad_por_particion, limites_particion = np.histogram(data, bins=particiones)
print(cantidad_por_particion)
```

# density and contour plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-notebook")

def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# when single color used, negatives values are represented with dashed lines; positive -> solid lines
plt.contourf(X, Y, Z, 20, cmap="Spectral") # 20 equally spaced intervals within the data range :: f -> filled
plt.colorbar()
```

### image representation

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='Spectral' )
plt.colorbar()
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=12)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy', alpha=0.5 )
plt.colorbar()
```

# subplots

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-white")

fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.8)

for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i) # rows, cols, index
    ax.text(0.5, 0.5, str((2, 3, i)), fontsize=18, ha='center')
    
plt.show()
```

```python
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')

for i in range(2):
    for j in range(3):
        ax[i, j].text(0.5, 0.5, str((i,j)), fontsize=18, ha='center')
```
