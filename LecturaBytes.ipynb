```python
# fil1test.bin
import struct
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

file_name = "fil1test.bin"
with open(file_name, 'rb') as f:
    x = f.read()
    valores = np.double(struct.unpack("d"*int(len(x)/8), x)) # size double = 8B

    plt.style.use('seaborn-notebook')
    fig = plt.figure()
    ax = plt.axes()

    ax.plot(valores, color="#2CB5D4", linestyle='-')
    plt.show()
```
